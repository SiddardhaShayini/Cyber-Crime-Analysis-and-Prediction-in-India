```python
# Exclude TOTAL rows
exclude_labels = ["TOTAL (STATES)", "TOTAL (UTS)", "TOTAL (ALL INDIA)"]
state_data = df[~df["State/UT"].str.strip().str.upper().isin(exclude_labels)]
```

```python
# Define years as strings
years = [str(year) for year in range(2002, 2022)]

# Plot all state/UT trends
plt.figure(figsize=(14, 8))

for _, row in state_data.iterrows():
    plt.plot(years, row[years].values.flatten(), label=row["State/UT"])

plt.xlabel("Year")
plt.ylabel("Cyber Crime Incidents")
plt.title("Yearly Cyber Crime Trends for All States/UTs (2002–2021)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', ncol=1)
plt.tight_layout()
plt.show()
```

```python
# Sort states by total
state_totals = state_data.sort_values(by="Total", ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(state_totals["State/UT"], state_totals["Total"])
plt.xlabel("State/UT")
plt.ylabel("Total")
plt.title("Total Values for States/UTs")
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 states based on total
top_states = state_totals.head(10)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_states["Total"], labels=top_states["State/UT"], autopct="%1.1f%%", startangle=140)
plt.title("Distribution of Total Values Across Top 10 States")
plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = state_data[years + ["Total", "Total-Scale"]].corr()

# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

```python
# Sort by Total-Scale
scale_data = state_data.sort_values(by="Total-Scale", ascending=False)

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(scale_data["State/UT"], scale_data["Total-Scale"], marker="o")
plt.xlabel("State/UT")
plt.ylabel("Total-Scale")
plt.title("Total-Scale Values Across States/UTs")
plt.xticks(rotation=90)
plt.show()
```

```python
# Transpose data to get years as index and states as columns
data = state_data.set_index("State/UT")[years].T

# Plot stacked bar chart
plt.figure(figsize=(16, 8))
data.plot(kind="bar", stacked=True, figsize=(16, 8), width=0.8)

plt.xlabel("Year")
plt.ylabel("Number of Cyber Crimes")
plt.title("Stacked Bar Chart of Yearly Cyber Crimes for All States/UTs")
plt.legend(title="State/UT", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Area plot for all states combined
total_yearly_data = state_data[years].sum()

plt.figure(figsize=(12, 6))
plt.fill_between(years, total_yearly_data, color="skyblue", alpha=0.5)
plt.plot(years, total_yearly_data, color="blue")
plt.xlabel("Year")
plt.ylabel("Cumulative Values")
plt.title("Area Plot of Cumulative Trends Over Years")
plt.show()
```

```python
import plotly.express as px

# Create Sunburst Chart
fig = px.sunburst(
    state_data,
    path=["State/UT"],
    values="Total",
    title="Sunburst Chart: Total Values by States/UTs",
)
fig.show()
```

```python
# Prepare cumulative data for all states
selected_states = state_data.copy()  # Use all states
selected_states = selected_states.set_index("State/UT")[years]  # Drop 'State/UT' into index and select year columns
selected_states = selected_states.T  # Transpose to have years as index and states as columns

plt.figure(figsize=(14, 7))

# Plot stacked area chart
plt.stackplot(
    selected_states.index,  # years
    selected_states.values.T,  # each state's values (rows → columns)
    labels=selected_states.columns  # each state/UT name
)

plt.xlabel("Years")
plt.ylabel("Values")
plt.title("Stacked Area Chart of Yearly Contributions (All States/UTs)")
plt.legend(loc="upper left", fontsize="small", ncol=4, bbox_to_anchor=(0, -0.2))
plt.tight_layout(rect=[0, 0.1, 1, 1])  # Leave space for legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Extract only the totals
totals = df[df["State/UT"].isin(["TOTAL (STATES)", "TOTAL (UTS)", "TOTAL (ALL INDIA)"])]
years = [str(year) for year in range(2002, 2022)]
```

```python
fig = go.Figure()
for state in totals["State/UT"]:
    fig.add_trace(go.Scatter(
        x=years,
        y=totals[totals["State/UT"] == state][years].values.flatten(),
        mode="lines",
        fill="tozeroy",
        name=state
    ))

fig.update_layout(title="Area Chart of Total Cyber Crimes Over Years",
                  xaxis_title="Year", yaxis_title="Number of Cases")
fig.show()
```

```python
fig = px.bar(
    totals,
    x="State/UT",
    y="2021",
    title="Cyber Crimes in 2021 – TOTAL STATES vs UTS vs ALL INDIA",
    color="State/UT"
)
fig.update_layout(xaxis_title="Category", yaxis_title="Cases in 2021")
fig.show()
```
